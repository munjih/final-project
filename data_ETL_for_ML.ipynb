```python
import pandas as pd
```

```python
# Import data file
data = pd.read_csv('US_counties_COVID19_health_weather_data.csv')
data.head()
```

```python
# Drop all columns with 'num_'
data = data[data.columns.drop(list(data.filter(regex='num_')))]
data.head()
```

```python
# Drop all weather columns
data = data.loc[:,:'percentile_rank_social_vulnerability']
data.head()
```

```python
# Select rows for most recent date
data_recent = data[data.date == data.date.max()]
data_recent.head()
```

```python
# Check columns with null values
data_recent.columns[data.isna().any()].tolist()
```
